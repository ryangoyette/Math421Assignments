````r
---
title: "Assignment 17"
author: "Ryan Goyette"
date: "November 19, 2018"
output: html_document
---

```{r}
##Read in data.
cars=read.csv("C:\\Users\\student\\Documents\\Senior\\R\\cars.csv", header=TRUE)

##Change variable names.
names(cars)[1]="region"
names(cars)[2]="statecase"
names(cars)[3]="vehnum"
names(cars)[4]="pernum"
names(cars)[5]="county"
names(cars)[6]="day"
names(cars)[7]="month"
names(cars)[8]="hour"
names(cars)[9]="minute"
names(cars)[10]="age"
names(cars)[11]="sex"
names(cars)[12]="pertype"
names(cars)[13]="target"
names(cars)[14]="seatpos"
names(cars)[15]="drinking"
names(cars)[16]="year"
names(cars)[17]="mancoll"
names(cars)[18]="owner"
names(cars)[19]="modelyear"
names(cars)[20]="speed"
names(cars)[21]="deformed"
names(cars)[22]="weekday"
names(cars)[23]="route"
names(cars)[24]="latitude"
names(cars)[25]="longitude"
names(cars)[26]="harmev"
names(cars)[27]="light"
names(cars)[28]="weather"

##Removing variables.
cars$statecase=NULL
cars$county=NULL
cars$year=NULL
cars$minute=NULL
cars$vehnum=NULL
cars$pernum=NULL
cars$latitude=NULL
cars$longitude=NULL

##Changing variable types.
cars$age=as.integer(cars$age)
cars$modelyear=as.integer(cars$modelyear)
cars$speed=as.integer(cars$speed)

##Reduce levels.
levels(cars$state)=c("Southeast", "Noncontiguous", "Southwest", "Southeast", "Pacific", "RockyMountains", "Northeast", "Southeast", "Southeast", "Southeast", "Southeast", "Noncontiguous", "RockyMountains", "Midwest", "Midwest", "Midwest", "Midwest", "Southeast", "Southeast", "Northeast", "Southeast", "Northeast", "Midwest", "Midwest", "Southeast", "Midwest", "RockyMountains", "Midwest", "RockyMountains", "Northeast", "Northeast", "Southwest", "Northeast", "Southeast", "Midwest", "Midwest", "Southwest", "Pacific", "Northeast", "Northeast", "Southeast", "Midwest", "Southeast", "Southwest", "RockyMountains", "Northeast", "Southeast", "Pacific", "Southeast", "Midwest", "RockyMountains")
levels(cars$sex)=c("Female", "Male", "Female", "Male")
levels(cars$pertype)=c("Cyclist", "Driver", "Occupant", "Occupant", "Cyclist", "Occupant", "Pedestrian", "Pedestrian", "Pedestrian", "Occupant", "Pedestrian")
levels(cars$target)=c("Death", "Death", "Survived", "Survived", "Survived", "Survived", "Survived", "Survived")
levels(cars$seatpos)=c("Fourth", "Fourth", "Fourth", "Fourth", "Fourth", "Front", "Front", "Front", "Front", "Front", "Other", "Other", "Other", "Other", "Other", "Other", "Second", "Second", "Second", "Second", "Second", "Other", "Third", "Third", "Third", "Third", "Third", "Other", "Other")
levels(cars$mancoll)=c("Not a Collision with Motor Vehicle In-Transport", "Angle", "Front-to-Front", "Front-to-Rear", "Not a Collision with Motor Vehicle In-Transport", "Other", "Other", "Other", "Other", "Sideswipe", "Sideswipe", "Other")
levels(cars$owner)=c("Other", "NotOwner", "Owner", "Parked", "Other", "Other", "NotOwner", "NotOwner", "NotOwner")
levels(cars$deformed)=c("Unknown", "Disabling Damage", "Functional Damage", "Minor Damage", "No Damage", "Unknown", "Unknown")
levels(cars$route)=c("Country Road", "Interstate", "Local Street", "Other", "Highway", "Highway", "Other")
levels(cars$light)=c("Dark", "Dark", "Dark", "Dawn", "Daylight", "Dusk", "Unknown", "Unknown", "Unknown")
levels(cars$weather)=c("Windy", "Clear", "Not-Clear", "Not-Clear", "Wet", "Unknown", "Unknown", "Wet", "Wet", "Wet", "Wet", "Unknown")
levels(cars$harmev)=c("Structure", "Structure", "Structure", "Structure", "Structure", "Structure", "Structure", "Structure", "Ditch", "Structure", "Ditch", "Ditch", "Fell", "Small/Weak Structure", "Small/Weak Structure", "Fire", "Ground", "Structure", "Structure", "Water", "Small/Weak Structure", "Non-Collision", "Jackknife", "Person/Animal", "Small/Weak Structure", "Vehicle", "Person/Animal", "Object Fell", "Structure", "Non-Collision", "Small/Weak Structure", "Small/Weak Structure", "Structure", "Vehicle", "Ground", "Person/Animal", "Person/Animal", "Vehicle", "Person/Animal", "Rollover", "Small/Weak Structure", "Structure", "Object Fell", "Small/Weak Object", "Small/Weak Object", "Structure", "Other", "Small/Weak Object", "Structure", "Vehicle")
levels(cars$harmev)=c("Structure", "Small/Weak Structure", "Other", "Small/Weak Structure", "Fire", "Small/Weak Structure", "Water", "Non-Collision", "Non-Collision", "Person/Animal", "Vehicle", "Other", "Rollover", "Small/Weak Structure", "Other")
levels(cars$harmev)=c("Structure", "Small/Weak Structure", "Other", "Other", "Other", "Non-Collision", "Person/Animal", "Vehicle", "Other")

##Removing Rows With Missing Data.
cars=cars[cars$speed<133,]
cars=cars[cars$modelyear>1,]
cars=cars[cars$age<103,]
cars=cars[cars$hour<25,]

##Visualization and Graphs
library(ggplot2)
ggplot(cars, aes(x=cars$day, color=cars$target))+geom_density()+labs(x="Day", color="Target")
###Day seems to have little relationship with the target.

ggplot(cars, aes(x=cars$hour, color=cars$target))+geom_density()+labs(x="Hour", color="Target")
###The likelihood of an accident happening at all increases quite a lot towards the end of the day.
###In the early hours of the day, the probability of dying in the accident is higher than surviving, though it switches at around hour 13.

ggplot(cars, aes(x=cars$age, color=cars$target))+geom_density()+labs(x="Age", color="Target")
###Before age 20 and between ages 26 and 49, surviving an accident is more likely than dying. Between ages 20 and 26, the likelihood of both are roughly equal. Death becomes more prevalent than survival after age 50.

ggplot(cars, aes(x=cars$modelyear, color=cars$target))+geom_density()+labs(x="Model Year", color="Target")
###Model years before 70 show almost equal levels of death and survival, though death is consistently slightly higher. Between 70 and 75, survival becomes much more prevalent before once again becoming roughly equal, though this time survival is slightly higher.

ggplot(cars, aes(x=cars$speed, color=cars$target))+geom_density()+labs(x="Speed", color="Target")
###Survival is consistently more likely at speeds of less than roughly 65, and less likely above.

ggplot(cars, aes(x=cars$region, fill=cars$target))+geom_bar()+labs(x="Region", fill="Target")
###The ratio of death to survival seems roughly equal in all of the regions except for the Pacific, which seems to have higher percent of survival. 

ggplot(cars, aes(x=cars$month, fill=cars$target))+geom_bar()+labs(x="Month", fill="Target")
###The ratio of death to survival seems roughly equal in all of the months. 
###The number of accidents is noticeably higher in August and July and lower in February than in other months. 

ggplot(cars, aes(x=cars$sex, fill=cars$target))+geom_bar()+labs(x="Sex", fill="Target")
###The proportion of males that die looks a little higher than the proportion of females that die.
###There are about twice as many males as females in the data.

ggplot(cars, aes(x=cars$pertype, fill=cars$target))+geom_bar()+labs(x="Person Type", fill="Target")
###A significantly higher proportion of drivers died in accidents than Occupants.

ggplot(cars, aes(x=cars$seatpos, fill=cars$target))+geom_bar()+labs(x="Seat Position", fill="Target")
###A significantly higher proportion of those in the front seat died than those in any other seat position. 

ggplot(cars, aes(x=cars$drinking, fill=cars$target))+geom_bar()+labs(x="Drinking", fill="Target")
###Alcohol involved or unknown had a far higher proportion of deaths than no alcohol or not reported.

ggplot(cars, aes(x=cars$mancoll, fill=cars$target))+geom_bar()+labs(x="Collision", fill="Target")
###Collisions that did not involve the vehicle hitting another vehicle resulted in a much higher proportion of death than other collision types.

ggplot(cars, aes(x=cars$owner, fill=cars$target))+geom_bar()+labs(x="Owner", fill="Target")
###Other has a very high proportion of death, though given the uncertainty of the level this does not seem overly important.

ggplot(cars, aes(x=cars$deformed, fill=cars$target))+geom_bar()+labs(x="Deformed", fill="Target")
###The deformed variable level Disabling Damage held almost all of the death in the data.

ggplot(cars, aes(x=cars$weekday, fill=cars$target))+geom_bar()+labs(x="Weekday", fill="Target")
###The days of the week do not show noticeable differences in proportion of death.

ggplot(cars, aes(x=cars$route, fill=cars$target))+geom_bar()+labs(x="Route", fill="Target")
###Different routes do not seem to show noticeable differences in proportion of death.

ggplot(cars, aes(x=cars$harmev, fill=cars$target))+geom_bar()+labs(x="Harmev", fill="Target")
###Hitting a structure caused a high chance of death, while hitting a person or animal had a very low chance of death.

ggplot(cars, aes(x=cars$light, fill=cars$target))+geom_bar()+labs(x="Light", fill="Target")
###The light condition does not seem to have much importance in determining the chance of death. There were many more accidents in the dark or daylight than other conditions, but that is possibly due to their relatively longer duration.

ggplot(cars, aes(x=cars$weather, fill=cars$target))+geom_bar()+labs(x="Weather", fill="Target")
###The weather doesn't seem to have much of an effect on survival chance. 

#Undersampling to create a balance dataset of 1000 observations.
library(ROSE)
carsunder <- ovun.sample(target ~ ., data = cars, method = "under", N = 19598, seed = 1)$data
carssmall=carsunder[sample(1:nrow(carsunder), 1000),]

#Train a random forest with 7-fold cross-validation.
library(caret)
library(ranger)
model17.1 <- train(target~.,data = carssmall, method = "ranger", 
               trControl = trainControl(method ="cv", number = 7, verboseIter = TRUE))
cm17.1=confusionMatrix(data = model17.1, reference = carssmall$target, positive = "2")
cm17.1
##Accuracy=.708, mtry=69, splitrule=extratrees, min.node.size=1

#Tuning
myGrid17.2 = expand.grid(mtry = c(1:4), splitrule = c("gini","extratrees"),
                     min.node.size = c(1:5))
model17.2 <- train(target~.,data = carssmall, method = "ranger", 
               trControl = trainControl(method ="cv", number = 7, verboseIter = TRUE),
               tuneGrid = myGrid17.2)
cm17.2=confusionMatrix(data = model17.2, reference = carssmall$target, positive = "2")
cm17.2
##Accuray=.727, mtry=3, splitrule=extratrees, min.node.size=2

#Default glmnet
mycontrol17.3<-trainControl(method="cv", number=7, summaryFunction=twoClassSummary, classProbs=TRUE, verboseIter=TRUE)
model7.3<-train(target~., carssmall, method="glmnet", trControl=mycontrol17.3)
cm17.3=confusionMatrix(data = model7.3, reference = carssmall$target, positive = "2")
cm17.3
##Accuracy=.719, alpha=.55, lambda=.0339

#Custom glmnet
myGrid17.4<-expand.grid(alpha=0:1, lambda=seq(.0001, 1, length=100))
model7.4<-train(target~., carssmall, method="glmnet", tuneGrid=myGrid17.4, trControl=mycontrol17.3)
cm17.4=confusionMatrix(data = model7.4, reference = carssmall$target, positive = "2")
cm17.4
##Accuracy=.711, alpha=1, lambda=.0304
```
````
