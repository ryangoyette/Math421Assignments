````r
#2)
##Loading in data set.
```{r}
adult=read.csv("C:\\Users\\student\\Documents\\Senior\\R\\adult.csv", header=FALSE)
```

#3)
##Changing variable names.
```{r}
names(adult)[1]="age"
names(adult)[2]="workclass"
names(adult)[3]="fnlwgt"
names(adult)[4]="education"
names(adult)[5]="educationnum"
names(adult)[6]="maritalstatus"
names(adult)[7]="occupation"
names(adult)[8]="relationship"
names(adult)[9]="race"
names(adult)[10]="sex"
names(adult)[11]="capitalgain"
names(adult)[12]="capitalloss"
names(adult)[13]="hoursperweek"
names(adult)[14]="nativecountry"
names(adult)[15]="target"
str(adult)
```

#4) & #5)
##Checking for missing data.
```{r}
sum(is.na(adult))
```

##Checking categorical variables for missing data and replacing it.
```{r}
levels(adult$workclass)
which.max(table(adult$workclass))
levels(adult$workclass)=c("Private", "Federal-gov", "Local-gov", "Never-worked", "Private", "Self-emp-inc", "Self-emp-not-inc", "State-gov", "Without-pay")
levels(adult$education)
levels(adult$maritalstatus)
levels(adult$occupation)
which.max(table(adult$occupation))
levels(adult$occupation)=c("Prof-specialty", "Adm-clerical", "Armed-Forces", "Craft-repair", "Exec-managerial", "Farming-fishing", "Handlers-cleaners", "Machine-op-inspct", "Other-service", "Priv-house-serv", "Prof-specialty", "Protective-serv", "Sales", "Tech-support", "Transport-moving")
levels(adult$relationship)
levels(adult$race)
levels(adult$sex)
levels(adult$nativecountry)
which.max(table(adult$nativecountry))
levels(adult$nativecountry)=c("United-States", "Cambodia", "Canada", "China", "Columbia", "Cuba", "Dominican-Republic", "Ecuador", "El-Salvador", "England", "France", "Germany", "Greece", "Guatemala", "Haiti", "Holand-Netherlands", "Honduras", "Hong", "Hungary", "India", "Iran", "Ireland", "Italy", "Jamaica", "Japan", "Laos", "Mexico", "Nicaragua", "Outlying-US(Guam-USVI-etc)", "Peru", "Philippines", "Poland", "Portugal", "Puerto-Rico", "Scotland", "South", "Taiwan", "Thailand", "Trinadad&Tobago", "United-States", "Vietnam", "Yugoslavia")
levels(adult$target)
levels(adult$target)=c("0","1")
```

##Checking nominal variables for missing data.
##Capital Gains looks odd, but there is no information on it so it must be left alone.
```{r}
summary(adult)
str(adult)
```

#6)
##Loading in ggplot and performing data visualization.
```{r}
library(ggplot2)
```

##There are more males than females in the data. The largest race in both genders is white followed by black. The percentage of each race in each gender looks roughly equal.
```{r}
ggplot(adult, aes(x=sex, fill=race))+geom_bar()
```

##There are many more husbands than wives in the data. Excepting the gender specific relationship levels, all of the relationship levels have a roughly even split between genders except for Unmarried which is composed mostly of males. 
```{r}
ggplot(adult, aes(x=relationship, fill=sex))+geom_bar()
```

##Roughly 25% of the people in the data have a wage greater than 50K. The percentage of those with a wage greater than 50K who are females is much smaller than the percentage of those with a wage less than 50K who are female.
```{r}
ggplot(adult, aes(x=target, fill=sex))+geom_bar()
```

##The distribtion of females in the data is younger than the distribution of males. The peak of the denisty plot for females is at roughly 22 years, while the peak of the density plot for males is at roughly 36 years. The density plot for males is also flatter than the density plot for females.
```{r}
ggplot(adult,aes(x=age, color=sex))+geom_density()
```

##The distribution of those with a wage greater than 50K is older than the distribution of those with a wage less than 50K. The peak of the density plot for those with a wage greater than 50K is from 37-45 years, while the peak of the density plot for those with a wage less than 50K is at 22 years.
```{r}
ggplot(adult,aes(x=age, color=target))+geom_density()
```

##The density plot of hours per week for males and the density plot of hours per week for females both have a peek at 40 hours, though the peak for females is lower. The density plot of hours per week for females is skewed towards fewer than 40 hours, while the density plot of hours per week for males is skewed towards more than 40 hours.
```{r}
ggplot(adult,aes(x=hoursperweek, color=sex))+geom_density()
```

##The density plot of hours per week for those with a wage greater than 50K and the density plot of hours per week for those with a wage less than 50K both have a peek at 40 hours, though the peak for those with a wage greater than 50K is lower. The density plot of hours per week for those with a wage less than 50K is skewed towards fewer than 40 hours, while the density plot of hours per week for those with a wage greater than 50K is skewed towards more than 40 hours.
```{r}
ggplot(adult,aes(x=hoursperweek, color=target))+geom_density()
```

#7)
##Partitioning the data.
```{r}
library(caret)
splitIndex <- createDataPartition(adult$target, p = .60, list = FALSE, times = 1)
train <- adult[ splitIndex,]
test <- adult[-splitIndex,]
```

#8)
##Loading rpart.
```{r}
library(rpart)
```

##Building tree.
```{r}
mytree <- rpart(target ~ ., data = train, method = "class")
```

##Predicting on testing data.
```{r}
pred <- predict(mytree,test, type = "class")
```

#9)
##Evaluating the predictions.
```{r}
cm=confusionMatrix(data = pred, reference = test$target, positive = "1")
cm
```

#10)
##Ploting the tree.
```{r}
library(rattle)
fancyRpartPlot(mytree)
```

#11)
##Building and evaluating tree model using caret.
```{r}
model <- train(target~.,data =train, method = "rpart")
pred2=predict(model,test)
```

##Confusion matrix.
```{r}
cm2=confusionMatrix(data = pred2, reference = test$target, positive = "1")
cm2
```

#12)
##Function to automate 7-10.
###n is partition percentage to train.
```{r}
func101=function(n)
{
splitIndex <- createDataPartition(adult$target, p = n, list = FALSE, times = 1)
train <- adult[ splitIndex,]
test <- adult[-splitIndex,]
mytree <- rpart(target ~ ., data = train, method = "class")
pred <- predict(mytree,test, type = "class")
cm=confusionMatrix(data = pred, reference = test$target, positive = "1")
print(cm)
fancyRpartPlot(mytree)
}
```

##n=.7
```{r}
func101(.7)
```

##n=.8
```{r}
func101(.8)
```

##n=.9
```{r}
func101(.9)
```
````
